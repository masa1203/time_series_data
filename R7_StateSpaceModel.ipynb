# 状態空間モデルとはなにか

- 「様々な分析手法を統合したフレームワーク」
- 「Box-Jenkins法と比べて属人性が高い」
- 「目に見えない【状態】があることを過程した時系列モデルの1種」
- 「状態」と「観測値」を切り分けることによって人間が理解しやすい形でデータを表現する

## 状態空間モデルのデータの表現

状態空間モデルは**状態方程式** と **観測方程式** でデータを表現する。

- 状態方程式: 状態の変化のプロセスを記述
- 観測方程式: 状態から観測地が得られるプロセスを記述

日本語で表現すると次のようになる。

$
\begin{align}
    状態 &= 前時点の状態を用いた予測値 + 過程誤差　\\
    観測地 &= 状態 + 観測誤差
\end{align}
$

## パラメタ推定

データの表現方法とパラメタ推定を分けて考えること。  
パラメタ推定方法は大きく2つある。
- 〇〇フィルタ(フィルタリング系)
    - カルマンフィルタ+最尤法、散漫カルマンフィルタ
* MCMC
    - ベイズ推論 + HMC法

## パラメタ推定: カルマンフィルタと最尤法

<デメリット>
- **正規分布に従い、かつ線形なデータにしか適用できない**
    - **線形ガウス状態空間モデル** または **動的線形モデル(Dynamic Linear Models : DLM)** と呼ばれる

<メリット>
- MCMCと比べると計算は比較的容易
- パラメタ推定と状態の推定を分けることができる。
    - 一度パラメタさえ決まれば、データが手にはいるたび、逐次的に状態の推定や予測を行うことができる。

## パラメタ推定: ベイズ推論とHMC法

<デメリット>
- 計算不可、属人性が高く分析に手間と時間がかかる。
- 状態の推定とパラメタの推定を同時に行うため、逐次的な処理は苦手。
- Stanというソフトを習熟する必要がある。

<メリット>
- 非正規分布、非線形のデータにも適用可能
- Stanを使ってデータ生成過程を直接コードに書き下すことができる
